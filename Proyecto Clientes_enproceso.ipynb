# **Exploración estadística del perfil de clientes de una entidad bancaria**


---



## **Descripción**
Las entidades bancarias, gracias a la mayor velocidad de procesamiento de las computadoras y el aumento continuo del tamaño de los dispositivos de almacenamiento de información, pueden agregar constantemente información de sus clientes en sus bases de datos. Estas bases de datos contienen información útil que permite reducir la incertidumbre a la hora de tomar decisiones.

En el ámbito bancario, utilizando estas bases de datos, diversos análisis son de interés. Por ejemplo:
* identificar patrones de uso de productos y servicios,
* agrupar clientes según sus características demográficas y financieras,
* predecir la probabilidad de venta de productos financieros,
* mejorar la efectividad de campañas de marketing basadas en el comportamiento del cliente,
* evaluar la estabilidad financiera y el riesgo crediticio de los clientes.

En este proyecto se analiza un conjunto de datos proporcionado por una entidad bancaria y se aplican técnicas que permitan clasificar aquellos clientes que tienen una inclinación alta hacia la adquisición de plazos fijos. De esta manera, dicha entidad podrá potenciar sus ventas de plazos fijos.


## **Objetivo**
Realizar un análisis predictivo y/o exploratorio exhaustivo con el propósito de discernir y clasificar aquellos clientes con una inclinación alta hacia la adquisición de plazos fijos.


## **Descripción de la base de datos**
Se cuenta con información de clientes de una entidad bancaria, desde marzo a julio del 2023.
Las variables registradas están relacionadas con la gestión y análisis de productos bancarios y el comportamiento financiero de los clientes.

### Variables generales:
* key_value: Número de documento encriptado del cliente (identificador cliente)
* codmes: Mes de evaluación (formato AAAAMM)
* short_date: Día de impresión (formato DD/MM/AAAA)
* dia: Día de la impresión (1 al 31)
* prox_15: Cercanía a la quincena del mes (1 al 15)
* day_week: Orden del día de la semana (1 al 7)
* target: Indicador de si la impresión generó una venta (0: no, 1: sí) (Variable respuesta)

### Información demográfica y de segmentación:
* rango_edad: Rango de edad utilizado para segmentar clientes (1 al 6)
* flagcy_sex: Flag género del cónyuge (Soltero/a-0, M-1, F-2)
* flag_civil: Estado civil (C-3, V-2, S-1)
* gency_m: Género del cónyuge masculino (0 y 1)
* flag_costa: Indicador de si el cliente vive en una zona costera (0: no, 1: sí)
* flag_frontera: Indicador de si el cliente vive en una zona fronteriza (0: no, 1: sí)

### Información de productos bancarios:
* productos: Total de productos del banco que posee el cliente
* saldo_prom_parcial: Saldo promedio de productos pasivos sin incluir plazo fijo (Saldo fin de período total pasivo sin tener en cuenta el saldo de plazo fijo)
* saldo_prom_tot_millo_premia: Saldo promedio de la cuenta millonaria premia (Saldo promedio total del producto cuenta millonaria premia)
* saldo_fdp_tot_txs: Saldo final de periodo de todas las transacciones
* saldo_fdp_tot_alcancia: Saldo final de periodo en productos de alcancía
* saldo_prom_tot_activo: Saldo promedio total de productos activos
* saldo_prom_tot_tc: Saldo promedio total de tarjetas de crédito
* saldo_prom_tot_extracash_tc: Saldo promedio total de extracash en tarjeta de crédito
* saldo_prom_tot_libre_dispon: Saldo promedio total de libre disponibilidad
* saldo_fdp_tot_activo: Saldo final de periodo total de productos activos
* saldo_fdp_tot_adel_sueldo: Saldo final de periodo de adelanto de sueldo
* saldo_fdp_tot_ahor_cred: Saldo final de periodo de ahorro-crédito
* saldo_fdp_tot_convenio: Saldo final de periodo en productos de convenio
* saldo_fdp_tot_cts: Saldo final de periodo en CTS (Compensación por Tiempo de Servicios)
* saldo_fdp_tot_extracash_tc: Saldo final de periodo de extracash en tarjeta de crédito
* saldo_fdp_tot_hipotecario: Saldo final de periodo de productos hipotecarios
* saldo_fdp_tot_hipotecario_cov19: Saldo final de periodo de productos hipotecarios reprogramados debido a COVID-19
* saldo_fdp_tot_inversion: Saldo final de periodo de productos de inversión
* saldo_fdp_tot_libre_dispon: Saldo final de periodo de libre disponibilidad
* saldo_fdp_tot_millo_premia: Saldo final de periodo de cuenta millonaria premia (Saldo fin de periodo de clientes con producto millonaria premia)
* saldo_fdp_tot_millo_super_tasa: Saldo final de periodo de cuenta millonaria super tasa
* saldo_fdp_tot_millonaria: Saldo final de periodo de cuentas millonarias
* saldo_fdp_tot_pasivo: Saldo final de periodo de productos pasivos
* saldo_fdp_tot_planilla: Saldo final de periodo de productos de planilla
* saldo_fdp_tot_plazo_fijo: Saldo final de periodo de plazo fijo

### Información de Antigüedad y Cantidad de Productos
* ant_1_prod_month: Antigüedad en meses del primer producto en el banco hace un mes
* ant_1_prod_day: Antigüedad en días del primer producto en el banco hace un mes
* ant_1_prodact_day: Antigüedad en días del primer producto activo retail en el banco hace un mes
* ct_prodact: Cantidad de productos activos retail en el banco hace un mes
* ant_1_prodpas_day: Antigüedad en días del primer producto pasivo retail en el banco hace un mes
* ct_prodpas: Cantidad de productos pasivos retail en el banco hace un mes
* ant_1_prodpas_month: Antigüedad en meses del primer producto pasivo retail en el banco hace un mes
* ant_1_prodact_month: Antigüedad en meses del primer producto activo retail en el banco hace un mes

### Información Adicional
* flag_bonificacion: Indicador de bonificación extra en el mes, esto es si en el mes actual el cliente percibe algun ingreso adicional por mes (ejemplo: utilidades, meses que se reciben los CTS y adicional, dos meses en que reciben los aguinaldos) (0: no, 1: sí)
* tenencia: Cantidad de productos sin extracash, pago automático y seguros
* prm_sowe1ptcallsf06m: Promedio del share of wallet en la entidad principal por tarjeta de crédito reportado por el sistema financiero en los últimos seis meses
* prm_usoep1tcallsf06m_ingb: Promedio del uso de tarjeta de crédito en la entidad principal reportado por el sistema financiero en los últimos seis meses (entre el ingreso bruto)
* prm_sowe1ptcallsf06m_ingb: Promedio del share of wallet en la entidad principal por tarjeta de crédito reportado por el sistema financiero en los últimos seis meses (entre el ingreso bruto)
* prm_usotcrrstsf06m_ingb: Promedio del porcentaje de uso de la tarjeta de crédito de otras entidades bancarias en los últimos seis meses (entre el ingreso bruto)
* prm_usotcribksf06m_ingb: Promedio del % de uso de la tarjeta de crédito del banco en los últimos 6 meses (entre el ingreso bruto)

### Observaciones:
- Se observan inconsistencias entre la variable ‘rango_edad’ y ‘edad_m’. Se decide trabajar con ‘edad_m’ y realizar una nueva categorización de los datos, eliminando ‘rango_edad’ y creando uno nuevo. Para ello, se realiza previamente una transformación de la variable ‘edad_m’ de manera de obtener las edades pero en años, y se la llama ‘edad_a’. A partir de la misma se crean las siguientes categorías: ‘0 a 17 años’, ‘18 a 27 años’, ‘28 a 45 años’, ‘46 a 65 años’, ‘66 a 72 años’ y ‘73 o más’. Luego, se elimina ‘edad_a’ ya que en el modelo se decide trabajar con los rangos.
-Se excluye la variable ‘short_date’, utilizada únicamente para quedarse con un único registro por cliente, ya que no es de interés para el modelo.
-Se elimina el campo ‘civ_s’, el cual indica si el cliente tiene estado civil soltero (1) o no (0). Esto se debe a que proporciona la misma información que la variable ‘flag_civil’, la cual indica el estado civil del cliente (1. Soltero, 2. Viudo, 3. Casado).
- Se elimina el campo ‘gency_m’, el cual indica si el cónyuge del cliente es de género masculino (1) o no (0). Dicho campo contiene la misma información que ‘flagcy_sex’, el cual indica si el cónyuge del cliente es de género masculino (1), femenino (2) o si no tiene (0).

```python
# Se eliminan variables que no van a usarse en el análisis
df2.drop('Conteo', axis=1, inplace=True)
df2.drop('rango_edad', axis=1, inplace=True) # Inconsistente con edad_m, se confía en los datos de edad_m
df2.drop('short_date', axis=1, inplace=True) # Se usa únicamente para quedarse con un único registro por cliente
df2.drop('civ_s', axis=1, inplace=True) # Misma información que flag_civil
df2.drop('gency_m', axis=1, inplace=True) # Misma información que flagcy_sex
```

Se toma como criterio eliminar aquellos campos que tienen un porcentaje de registros faltantes mayor al 50%

```python
df2.drop('prm_sowe1ptcallsf06m', axis=1, inplace=True)
df2.drop('prm_usoep1tcallsf06m_ingb', axis=1, inplace=True)
df2.drop('prm_sowe1ptcallsf06m_ingb', axis=1, inplace=True)
df2.drop('prm_usotcrrstsf06m_ingb', axis=1, inplace=True)
df2.drop('prm_usotcribksf06m_ingb', axis=1, inplace=True)
df2.drop('ant_1_prodact_day', axis=1, inplace=True)
df2.drop('ant_1_prodact_month', axis=1, inplace=True)
```

```python
# Se verifica que se hayan eliminado los campos con más del 50% de registros faltantes
df2.isnull().sum()
```

Para tratar los valores faltantes se decide imputar por la moda o la mediana. En el caso de variables cualitativas, se utiliza la moda. En caso de variables cuantitativas, se utiliza la mediana, ya que las distribuciones son asimétricas.

```python
# Variables cuantitativas
df2['ct_prodpas'].fillna(df2['ct_prodpas'].median(), inplace=True)
df2['ct_prodact'].fillna(df2['ct_prodact'].median(), inplace=True)
df2['ant_1_prodpas_day'].fillna(df2['ant_1_prodpas_day'].median(), inplace=True)
df2['ant_1_prodpas_month'].fillna(df2['ant_1_prodpas_month'].median(), inplace=True)
df2['ant_1_prod_month'].fillna(df2['ant_1_prod_month'].median(), inplace=True)
df2['ant_1_prod_day'].fillna(df2['ant_1_prod_day'].median(), inplace=True)
df2['edad_m'].fillna(df2['edad_m'].median(), inplace=True)
```

```python
# Variables cualitativas
moda_rango_ingreso = df2['rango_ingreso'].mode()[0]
df2['rango_ingreso'].fillna(moda_rango_ingreso, inplace=True)
```

```python
# Se verifica que no queden registros con valores faltantes
df2.isnull().sum()
```

Se crea un nuevo campo que contenga las edades en años, luego se realiza la categorización correspondiente.

```python
df2['edad_a'] = np.floor(df2['edad_m'] / 12).astype(int)
```

```python
# Creo la variable rango_edad
# Definir los límites de los rangos de edad
bins = [0, 17, 27, 45, 65, 72, float('inf')]
labels = ["0 a 17", "18 a 27", "28 a 45", "46 a 65", "66 a 72", "73 o más"]

# Crear la nueva columna usando pd.cut
df2['rango_edad '] = pd.cut(df2['edad_a'], bins=bins, labels=labels, right=True)
```

```python
# Trabajo con rango_edad
df2.drop('edad_m', axis=1, inplace=True)
df2.drop('edad_a', axis=1, inplace=True)
```

### Cercanía a la quincena del mes vs target

```python
# Calcular el total de clientes en cada categoría de prox_15
total_prox15 = df2.groupby('prox_15')['target'].count()

# Crear un DataFrame con la cantidad de clientes según target y para cada prox_15
df_target_por_prox15 = df2.groupby(['prox_15', 'target']).size().unstack(fill_value=0)

# Dividir los valores por el total y multiplicar por 100 para obtener el porcentaje
df_porcentaje_target_por_prox15 = df_target_por_prox15.div(total_prox15, axis=0) * 100

# Graficar el gráfico de barras
plt.figure(figsize=(10,6))
ax = df_porcentaje_target_por_prox15.plot(kind='bar', stacked=True)
plt.title('Distribución de target según prox_15')
plt.xlabel('Cercanía a la quincena del mes')
plt.ylabel('Porcentaje')
plt.xticks(rotation=0)
# Mover la leyenda al costado y fuera del gráfico
plt.legend(title='Adquisición plazo fijo', labels=['No', 'Sí'], bbox_to_anchor=(1.05, 1), loc='upper left')

# Agregar etiquetas de porcentaje dentro de cada barra
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.2f}%', (x + width/2, y + height/2), ha='center', va='center', fontsize=6)

plt.show()
```

## Modelos de clasificación
Se pretende ajustar un modelo que relacione la respuesta con los predictores de manera que sea posible entender mejor la asociación entre estas variables y/o predecir correctamente la respuesta para observaciones futuras.

Se decide trabajar con modelos de regresión logística y árboles de decisión, los cuales permiten predecir la probabilidad de que un cliente adquiera un plazo fijo basado en sus características generales y demográficas, en la información de los productos bancarios que posee y su comportamiento financiero.


#### División de la base en entrenamiento y prueba
Con la base de entrenamiento se construye y ajusta el modelo, mientras que con la de prueba se realiza la predicción y se evalúa su rendimiento.

Este procedimiento permite medir el rendimiento del modelo en datos no considerados durante el entrenamiento, asegurando una evaluación más realista y objetiva de su capacidad de generalización.

En este caso, se extrajo una muestra al azar del 70% de los clientes para crear la base de entrenamiento. El 30% restante conforman la base de prueba.


```python
from sklearn.model_selection import train_test_split
```

```python
# Defino las variables explicativas
X= df_final.drop(columns=['target'])
X
```

```python
# Defino variable respuesta
y= df_final['target']
y
```

```python
# Dividir los datos en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Se mantienen las proporciones de cada clase gracias a stratify=y
values_counts_y_train = np.asarray(np.unique(y_train, return_counts=True))
values_counts_y_train
```

### Regresión logística
La regresión logística modela la probabilidad de que una observación pertenezca a una de dos clases posibles. En este caso, el modelo se entrena para predecir la probabilidad de que un cliente adquiera un plazo fijo, estimando la relación entre las características de los clientes y su probabilidad de adquirir un plazo fijo.


Se ajusta el modelo de regresión logística utilizando una técnica de sobremuestreo y submuestreo que genera observaciones sintéticas a modo de balancear las clases (SMOTE).



```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
# Modelo regresión logística
classifier = LogisticRegression()
```

```python
# Aplicar SMOTE en el conjunto de entrenamiento
smote = SMOTE(k_neighbors=5 ,random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Comprobemos que las clases estan balanceadas
values_counts_SMOTE = np.asarray(np.unique(y_train_smote, return_counts=True))
values_counts_SMOTE
```

```python
# Entrenar un clasificador en los datos de entrenamiento con SMOTE (se ajusta el modelo con datos balanceados)
modelo_datos_SMOTE = classifier.fit(X_train_smote, y_train_smote)

# Realizar predicciones en el conjunto de prueba
y_pred_SMOTE = modelo_datos_SMOTE.predict(X_test)
```

### Matriz de confusión

- FALSE POSITIVE (error tipo I): 4.451 -> Hay 4.451 clientes que no habían adquirido el plazo fijo y el modelo predice que si
- FALSE NEGATIVE (error tipo II): 588 -> Hay 588 clientes que habían adquirido el plazo fijo y el modelo predice que no

F1-score clase 1: 0.18. El modelo tiene un rendimiento muy bajo para la clase que adquiere el plazo fijo. El modelo tiene dificultades para capturar los clientes que adquieren el plazo fijo correctamente.

El modelo está sesgado hacia la clase mayoritaria.

### Área bajo la curva (AUC/ROC)

La curva ROC traza la tasa de verdaderos positivos (recall) frente a la tasa de falsos positivos (1 - especificidad) a diferentes umbrales de decisión.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot

# Predecimos las probabilidades de que cada observación en el set de entrenamiento pertenezca a cada clase 0/1.
lr_probs = modelo_datos_SMOTE.predict_proba(X_train)
# Nos quedamos con las probabilidades de la clase positiva (la probabilidad de 1)
lr_probs = lr_probs[:, 1]
```

```python
#Generamos un clasificador sin entrenar que siempre asigna 0
ns_probs = [0 for _ in range(len(y_train))]
```

```python
# Calculamos el AUC
ns_auc = roc_auc_score(y_train, ns_probs) # AUC del clasificador ficticio, como siempre predice 0 es 0.5
lr_auc = roc_auc_score(y_train, lr_probs) # AUC del modelo de regresión logística
```

```python
# Imprimimos en pantalla
print('Sin entrenar: ROC AUC=%.3f' % (ns_auc))
print('Regresión Logística: ROC AUC=%.3f' % (lr_auc))
```

El AUC es del 74.2%, lo cual indica que el modelo tiene un desempeño moderado al discriminar entre las dos clases.

```python
# Calculamos las curvas ROC
# Calculo tasas de verdaderos positivos (TPR) y falsos positivos (FPR) para distintos umbrales, tanto para el clasificador ficticio como para el modelo de regresión logística
ns_fpr, ns_tpr, _ = roc_curve(y_train, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_train, lr_probs)
```

```python
# Grafico las curvas ROC
pyplot.plot(ns_fpr, ns_tpr, linestyle='--', label='Sin entrenar')
pyplot.plot(lr_fpr, lr_tpr, marker='.', label='Regresión Logística')
# Etiquetas de los ejes
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
pyplot.legend()
pyplot.show()
```

```python
# Calcular la curva ROC
fpr, tpr, umbrales = roc_curve(y_train, lr_probs)

# Calcular la distancia al punto (0, 1) para encontrar el punto de corte óptimo
distancias = np.sqrt((1 - tpr)**2 + fpr**2)
indice_optimo = np.argmin(distancias)
umbral_optimo = umbrales[indice_optimo]

print("Punto de corte óptimo:", umbral_optimo) # umbral óptimo para separar entre clases
```

```python
# Crear un DataFrame con los valores de fpr, tpr y umbrales
import pandas as pd
data = pd.DataFrame(zip(fpr, tpr, umbrales), columns=["FPR", "TPR", "Umbrales"])
data
```

```python
# Establecer un umbral para convertir las probabilidades en etiquetas binarias
# Si la probabilidad es mayor o igual al umbral la etiqueta será 1; si es menor, será 0.
umbral = 0.38714795164910293
etiquetas_predichas = (lr_probs >= umbral).astype(int)

etiquetas_predichas
```

### Árbol de decisión
El árbol de decisión utiliza una estructura jerárquica de decisiones para predecir la clase de una observación. Dividen el espacio de las variables explicativas en múltiples secciones o estratos, a partir de las cuales se construyen diferentes perfiles de individuos. Es decir, de igual manera que la regresión logística, la serie de decisiones para predecir si adquirirán un plazo fijo se basa en las características de los clientes.
